##### 예제 3.2
###### 1. 문제정의 : 리스트에 n개의 항목들이 들어 있다. 이 리스트에서 "탐색키"를 가진 항목을 찾아라. 만약 찾는 항목이 있으면 그 항목의 인덱스를 반환하고, 없으면 -1을 반환하라. 단, 리스트의 항목들은 정렬되어 있지 않다.
###### 2. 알고리즘 설명 : 입력 리스트의 첫 항목부터 순서대로 하나씩 탐색 키와 비교한다.
###### 3. 손으로 푼 예제 : 리스트 A에서 입력된 key값을 찾아 key값이 위치한 인덱스를 반환시킨다.
###### 4. 알고리즘 개요(입력 변수, 출력, 알고리즘 개요) : 
###### 4.1. 입력 변수 : [5, 3, 8, 4, 2, 7] => key값을 찾아볼 리스트
###### 4.2. 출력 : key값이 리스트에 있다면 키 값의 인덱스 출력, 없다면 없다는 문구 출력
###### 4.3. 알고리즘 개요 : 순차 탐색
###### 7. 수행 결과 : 
###### 8. 복잡도 분석 : for문으로 A의 크기 만큼 반복되는 것이므로 복잡도는 O(n)라고 할 수 있다.

```python
def sequential_search(A, key):
    for i in range(len(A)):
        if A[i] == key :
            return i
    return -1

data = [5, 3, 8, 4, 2, 7]
print("지정된 리스트 : ", data)
key = int(input("원하는 키 값을 입력하시오(리스트 속에 있어야 한다) >> "))
if(sequential_search(data,key) == -1):
    print("키 값이 리스트에 없습니다.")
else :
    print("키 값의 인덱스는 ", sequential_search(data, key), "입니다.")
```
