##### 예제 3.4
###### 문제정의 : 2차원 평면상에 n개의 점이 있다. 가장 인접한 쌍의 거리를 구하라.
###### 알고리즘 설명 : 가능한 모든 점의 쌍 (pi, pj)에 대해 거리를 계산하고, 가장 짧은 것을 찾는다.
###### 손으로 푼 예제 : 거리를 계산하며 변수에 값을 저장하고 저장된 값과 계산된 값들로 가장 짧은 거리를 비교하여 가장 짧은 거리 값을 리턴해준다. 따라서 리턴된 최근접 거리를 출력한다.
###### 알고리즘 개요(입력 변수, 출력, 알고리즘 개요) : 
###### 입력 변수 : 계산할 점들의 위치를 입력한다 Ex, (21, 34)
###### 출력 : 가장 가까운 거리를 출력한다
###### 알고리즘 개요 : 최근접 쌍의 거리
###### 7. 수행 결과 : 
###### 8. 복잡도 분석 : 이중 for문이 있어 O(n^2)의 복잡도를 가지고 있을 것 같고, 실제로는 n-1번 반복하는 for문 하나, i+1번째에서 n까지 반복하는 for문 하나이기에 (n-1)(n-(i+1))이므로 가장 큰 영향을 주는 것은 n^2이 맞다. 따라서 복잡도는 O(n^2)이다.
#### (교재 복잡도 분석) : ![3_Quest4 O(n^2).png](<attachment:3_Quest4 O(n^2).png>)

```python
import math

def distance(p1, p2):
    return math.sqrt((p1[0]-p2[0])**2 + (p1[1]-p2[1])**2)

def closet_pair(p):
    n = len(p)
    mindist = float("inf")
    for i in range(n-1):
        for j in range(i+1, n):
            dist = distance(p[i], p[j])
            if dist < mindist:
                mindist = dist
    return mindist

p = [(2,3), (12,30), (40,50), (5,1), (12,10), (3,4)]
print("최근접 거리:", closet_pair(p))
```
