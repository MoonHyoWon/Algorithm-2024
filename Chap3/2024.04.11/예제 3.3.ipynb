##### 예제 3.3
###### 1. 문제정의 : 길이가 n인 입력 문자열 T와 길이가 m인 패턴 문자열 P가 있다. T에서 가장 먼저 나타나는 P의 위치를 찾아라. 패턴이 없으면 -1을 반환하라.
###### 2. 알고리즘 설명 : 텍스트의 첫 번째 문자 위치에 패턴을 놓고 비교한다. 이 과정을 패턴을 오른쪽으로 한 칸씩 옮기면서 성공한 매칭이 나타날 때까지 반복한다.
###### 3. 손으로 푼 예제 : text를 알파벳 한 글자로 쪼개며, pattern과 비교하며 일치하는지 찾는다. 만약 입력된 패턴과 일치한 text 문구를 찾았다면, 기준이 되는 알파벳의 인덱스를 리턴한다. (Ex, pattern 값이 LO라면 일치하는 L의 인덱스를 리턴한다)
###### 4. 알고리즘 개요(입력 변수, 출력, 알고리즘 개요) : 
###### 4.1. 입력 변수 : 패턴 문자열(ex, LO)
###### 4.2. 출력 : 패턴 문자열 in 리스트 --> 위치
###### 4.3. 알고리즘 개요 : 문자열 매칭(억지 기법)
###### 7. 수행 결과 : 
###### 8. 복잡도 분석 : for문 안에 while문이 존재하여 O(n^2)의 복잡도를 가질 것 같지만, while문의 경우 패턴의 길이 대로 반복하기 때문에 (문제에선) 2정도의 반복을 한다. 따라서 O(n)의 복잡도를 가질 것이다.

```python
def string_matching(T, P):
    n = len(T)
    m = len(P)
    for i in range(n-m+1):
        j = 0
        while j < m and P[j] == T[i+j]:
            j = j+1
        if j == m:
            return i
    return -1

text = 'HELLO WORLD'
pattern = 'LO'
print(pattern, 'in', text, '-->', string_matching(text, pattern))
pattern = 'HI'
print(pattern, 'in', text, '-->', string_matching(text, pattern))
```
