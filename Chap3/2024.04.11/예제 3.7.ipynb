##### 예제 3.7 BFS
###### 문제정의 : 2차원 평면상에 n개의 점이 있다. 가장 인접한 쌍의 거리를 구하라.
###### 알고리즘 설명 : 가능한 모든 점의 쌍 (pi, pj)에 대해 거리를 계산하고, 가장 짧은 것을 찾는다.
###### 손으로 푼 예제 : 방문한 곳에 시작점을 추가한다. 큐에 시작점을 넣고, 큐에서 값을 얻어 출력한다. while문에선 방문한 곳에 u의 값을 넣고 큐에서 u를 출력한다. 큐가 비워질 때까지 계속 반복한다.
###### 알고리즘 개요(입력 변수, 출력, 알고리즘 개요) : 
###### 입력 변수 : 시작 점들과 연결된 점을 입력한다. (시작점 : {"도착점1", "도착점2"} 등의 딕셔너리를 사용한다)
###### 출력 : 한번에 이어진 채로 그릴 수 있는 방법이 출력된다.
###### 알고리즘 개요 : 그래프 탐색 (너비 우선 탐색/BFS)
###### 7. 수행 결과 : ![3_Quest6 BFS.png](<attachment:3_Quest6 BFS.png>)
###### 8. 복잡도 분석 : while문이 큐가 비워질 때까지 반복되고, for문도 u가 큐값에 있는 것까지 반복되기 때문에 복잡도는 O(n^2)일 것이다.

```python
import queue
def bfs(graph, start):
    visited = { start }
    que = queue.Queue()
    que.put(start)
    while not que.empty():
        v = que.get()
        print(v, end= " ")
        nbr = sorted(graph[v] - visited)
        for u in nbr:
            visited.add(u)
            que.put(u)

mygraph =  {"A":{"B", "C"}, "B":{"A", "D"}, "C":{"A", "D", "E"}, "D":{"B", "C", "F"}, "E":{"C", "G", "H"}, "F":{"D"}, "G":{"E", "H"}, "H":{"E", "G"}}

print('BFS : ', end= " ")
bfs(mygraph, "A")
print()
```
