##### 1) 문제 정의 (MD) : 공간을 희생하여 시간 효율을 높이고자 하는 방법
##### 2) 알고리즘 설명 (MD) : 항목들을 킷값에 따라 분배하여 저장한다. 자릿수의 값에 따라서 정렬하기 때문에 기수 정렬이다. 기수 정렬은 다단계 정렬인데, 단계의 수는 데이터의 전체 자릿수와 일치한다.
###### 1. 항목들을 저장할 버킷들을 준비한다.
###### 2. 입력 항목들을 순서대로 킷값에 따라 해당 버킷에 넣는다.
###### 3. 위쪽 버킷부터 순차적으로 버킷 안에 들어 있는 숫자를 출력한다.
##### 3) 손으로 푼 예제 (MD, 이미지 삽입) : ![AL6.1_radix_sort_손으로 푼 예제.jpg](<attachment:AL6.1_radix_sort_손으로 푼 예제.jpg>)
##### 4) 코드 개요(입력 변수, 출력, 함수 설명) (MD) : 
###### 입력 변수 : 리스트 A
###### 출력 변수 : 정렬된 리스트 A
###### 함수 설명 : 버킷을 통한 리스트 A를 분배하여 저장한 후 저장한 값을 다시 순차적으로 저장하며 정렬을 진행하는 함수이다.


```python
# 5) 코드 (PY,C)
BUCKETS = 10
DIGITS = 4

from queue import Queue
def radix_sort(A):
    queues = []
    for i in range(BUCKETS):
        queues.append(Queue())
    n = len(A)
    factor = 1
    for d in range(DIGITS):
        for i in range(n):
            queues[(A[i]//factor) % BUCKETS].put(A[i])
        j = 0
        for b in range(BUCKETS):
            while not queues[b].empty():
                A[j] = queues[b].get()
                j += 1
        factor *= BUCKETS
        print("step", d+1, A)
```

```python
# 6) 테스트 코드 (PY,C) , 입력을 바꾸어서
import random
data = []

for i in range(10):
    data.append(random.randint(1, 9999))
radix_sort(data)
print("Radix: ", data)
```

##### 7) 수행 결과 (MD, 결과 캡춰하여 이미지로 삽입) : 
##### 8) 복잡도 분석 (MD) : O(n)이다. 반복하여 버킷에 저장하는 연산이 리스트 A의 크기인 n번만큼 진행되며, 만약 한자릿수가 아니더라도 n+...+n 번이 진행되기에 최종적으로 kn번이 반복될 예정이다. (k는 상수) 여기서 k는 유의미한 변화를 주지 못하므로 복잡도는 O(n)이 되는 것이 옳다.
